```python
from sklearn.datasets import load_digits # 글씨체 데이터셋 가져오기
from sklearn.model_selection import train_test_split #트레이닝, 테스트 가져오기
import numpy as np #넘피 가져오기
import pandas as pd #판다스 가져오기
import matplotlib.pyplot as plt # 매트플롯립 가져오기


from sklearn.tree import DecisionTreeClassifier # 트리 트리거 가져오기
from sklearn.ensemble import RandomForestClassifier # 랜덤 트리거 가져오기
from sklearn.svm import SVC #SVC 트리거 가져오기
from sklearn.linear_model import SGDClassifier #SGD트리거 가져오기
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report # 머신러닝 학습표 가져오기 

```

## 1-2)데이터 준비하기

```python
digits=load_digits()

digits_data = digits.data #디지털 데이터셋에 데이터들을 변수에 선언

digits_label = digits.target #디지털 데이터셋에 타깃들을 변수에 선언

digits_feature_name=digits.feature_names #디지털 데이터셋에 모양새들을 선언

digits
```

___________________________________________________________________________________ 
## 1-4)Train,Test 데이터 분리하기

```python
test_size_list=list(range(20,40))
a= [i/100 for i in  test_size_list] #0.2~0.4까지 변수에 저장

for i in a:
#트레인과 테스트 데이터셋 분리
    X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                        digits_label, 
                                                        test_size=i, #0.2~0.4까지 20개를 본다
                                                        random_state=7)
```

### 나는 test_size값을 0.2~0.4까지 설정해 각 test_size값마다 훈련결과를 보고싶어 반복문을 사용해서 결과값을 도출해보았다.

## Decision Tree 사용해 보기

```python
test_size_list=list(range(20,40))
a= [i/100 for i in  test_size_list] #0.2~0.4까지 변수에 저장

for i in a:
#트레인과 테스트 데이터셋 분리
    X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                        digits_label, 
                                                        test_size=i, #0.2~0.4까지 20개를 본다
                                                        random_state=7)
    decision_tree = DecisionTreeClassifier(random_state=32)
    decision_tree.fit(X_train, y_train) #모델학습
    y_pred = decision_tree.predict(X_test)
    print(classification_report(y_test, y_pred))
```

### 20개의 결과를 비교해본 결과 test_size가 0.21로 설정해줬을때 정확도가 가장 좋게 나왔다. 하지만 다른것과 다르게 그렇게 큰 차이는 보이지 않는다.

## Random Forest 사용해 보기

```python
test_size_list=list(range(20,40))
a= [i/100 for i in  test_size_list] #0.2~0.4까지 변수에 저장

for i in a:
#트레인과 테스트 데이터셋 분리
    X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                        digits_label, 
                                                        test_size=i, #0.2~0.4까지 20개를 본다
                                                        random_state=7)
    random_tree = RandomForestClassifier(random_state=32)#랜덤스테이트르르 32로 설정
    random_tree.fit(X_train,y_train)# 머신러닝 학습
    y_pred = random_tree.predict(X_test)# 머신러닝 테스트
    print(classification_report(y_test, y_pred))#결과표 출력
```

### Random Forest를 사용한 결과로 정확도는 0.97의 결과값으로 가장 크게 나오며, Dicision Tree 보다 정확도가 높게 나온다.
